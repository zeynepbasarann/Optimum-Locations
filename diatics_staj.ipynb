```python
import matplotlib.pyplot as plt
import random
import math
```

```python
class Hayvan:
    hayvan_isimleri=[]
    def __init__(self, name, disi, hareket, erkek=None):
        self.name = name
        Hayvan.hayvan_isimleri.append(self.name)
    
        self.x_disi = [random.randint(0, 500) for i in range(disi)]
        self.y_disi = [random.randint(0, 500) for i in range(disi)]
        if erkek is not None:
            self.x_erk = [random.randint(0,   500) for i in range(erkek)]
            self.y_erk = [random.randint(0, 500) for i in range(erkek)]
        else:
            self.x_erk = []
            self.y_erk = []
        self.hareket = hareket
        

    def Hareket(self):
        yonler = ["yukari", "asagi", "saga", "sola"]
        for i in range(len(self.x_disi)):
          yon = random.choice(yonler)
          # Fixed indentation here
          if yon == "yukari":
              self.y_disi[i] += self.hareket
          elif yon == "asagi":
              self.y_disi[i] -= self.hareket
          elif yon == "saga":
              self.x_disi[i] += self.hareket
          else:
            self.x_disi[i] -= self.hareket
        if len(self.x_erk) > 0:
          for i in range(len(self.x_erk)):
              yon = random.choice(yonler)
              if yon == "yukari":
                  self.y_erk[i] += self.hareket
              elif yon == "asagi":
                  self.y_erk[i] -= self.hareket
              elif yon == "saga":
                  self.x_erk[i] += self.hareket
              else:
                  self.x_erk[i] -= self.hareket
      

    @staticmethod
    def uzaklik_hesapla(x1, y1, x2, y2):
        return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        
        
    def Avlanma(avci,av,mesafe):
        for i in range(len(avci.x_disi)):
            for j in range(len(av.x_disi)):
                uzaklik=Hayvan.uzaklik_hesapla(avci.x_disi[i],avci.y_disi[i], av.x_disi[j],av.y_disi[j])
                if uzaklik == mesafe:
                    del av.x_disi[i]
                    del av.y_disi[i]
                    break
            if len(av.x_erk)>0:
                for j in range(len(av.x_erk)):
                    uzaklik=Hayvan.uzaklik_hesapla(avci.x_erk[i],avci.y_erk[i], av.x_erk[j],av.y_erk[j])
                    if uzaklik == mesafe:
                        del av.x_erk[i]
                        del av.y_erk[i]
                        break
    
                    



    def Ureme(hayvan):
        for i in range(len(avci.x_disi)):
            for j in range(len(av.x_erk)):
                uzaklik=Hayvan.uzaklik_hesapla(avci.x_disi[i],avci.y_disi[i], av.x_erk[j],av.y_erk[j])
                if uzaklik == 3:
                    cinsiyet = random.choice(["dişi", "erkek"])
                    if cinsiyet == "dişi":
                        hayvan.x_disi.append(random.randint(0, 500))
                        hayvan.y_disi.append(random.randint(0, 500))
                    else:
                        hayvan.x_erk.append(random.randint(0, 500))
                        hayvan.y_erk.append(random.randint(0, 500))

                        
                    
```

```python
koyun=Hayvan("koyun",15,2,15)
kurt=Hayvan("kurt",5,3,5)
tavuk=Hayvan("tavuk",10,1)
horoz=Hayvan("horoz",10,1)
aslan=Hayvan("aslan",4,4,4)
avci=Hayvan("avci",1,1)
inek=Hayvan("inek",10,2,10)
```

```python
for i in range(1000):
    Hayvan.Avlanma(kurt,koyun,4)
    Hayvan.Avlanma(kurt,tavuk,4)
    Hayvan.Avlanma(kurt,horoz,4)
    Hayvan.Avlanma(aslan,inek,5)
    Hayvan.Avlanma(aslan,koyun,5)
    Hayvan.Avlanma(avci,koyun,8)
    Hayvan.Avlanma(avci,inek,8)
    Hayvan.Avlanma(avci,aslan,8)
    Hayvan.Avlanma(avci,tavuk,8)
    Hayvan.Avlanma(avci,horoz,8)
    Hayvan.Avlanma(avci,kurt,8)
    Ureme(kurt)
    Ureme(koyun)
    Ureme(aslan)
    Ureme(inek)
    koyun.Hareket()
    kurt.Hareket()
    inek.Hareket()
    horoz.Hareket()
    tavuk.Hareket()
    aslan.Hareket()
    
```
